```python
import json
```

```python
experiment_sets = {
    'original': ['vanilla'],
    'framing': ['neutral_framing', 'positive_framing', 'negative_framing'],
    'temporal': ['adv_past', 'adv_present', 'adv_future'],
    'perspective': ['adv_perspective_shift'],
    'quantifier': ['adv_inc', 'adv_dec']
}
```

```python
set_results = {}
```

```python
for experiment_set in experiment_sets:
    set_results[experiment_set] = {}
    for experiment in experiment_sets[experiment_set]:
        religion_results = json.load(open(f'./mlm_analysis/{experiment}/final_results/stereotype_tokens_religion_results.json'))
        for religion in religion_results:
            if religion_results[religion]:
                if religion not in set_results[experiment_set]:
                    set_results[experiment_set][religion] = {}
                set_results[experiment_set][religion][experiment] = []
                for subcategory in religion_results[religion]:
                    set_results[experiment_set][religion][experiment].extend(religion_results[religion][subcategory])
```
